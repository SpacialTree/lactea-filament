```python
cat_use = make_cat_use()
cat_raw = make_cat_raw()
```

```python
mask_qf = cat_raw.get_qf_mask()
mask_count = cat_raw.get_count_mask()
mask_brights = cat_raw.get_brights_mask()

mask = np.logical_and(mask_qf, mask_count)
mask = np.logical_and(mask, mask_brights)

cat_raw_masked = JWSTCatalog(cat_raw.catalog[mask])
```

```python
bins = np.linspace(13, 24, 50)
h_use, bins_use = np.histogram(cat_use.band('f182m'), bins=bins)
h_raw, bins_raw = np.histogram(cat_raw_masked.band('f182m'), bins=bins)

plt.step(bins_use[:-1], h_use, where='post', label='Multi-Detections Masked', alpha=0.5)
plt.step(bins_raw[:-1], h_raw, where='post', label='Regular Masks', alpha=0.5)
plt.xlabel('F182M Magnitude')
plt.ylabel('Count')
plt.yscale('log')
plt.legend()
```
